```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#sample data
np.random.seed()
data=np.random.normal(loc=50, scale=10, size=500)

#Histogram and KDE
plt.figure(figsize=(10,8))
sns.histplot(data, kde=True, color='g', bins=20)
plt.title('Distribution of Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#sample data
data = pd.DataFrame(
    {
        'category' : ['A', 'B', 'C', 'D'] * 20,
        'values' : np.random.randint(10, 50, 80)
    }
)
```

```python
#Box Plot
plt.figure(figsize=(10, 8))
sns.boxplot(x='category', y='values', data=data, palette='Set2')
plt.title('Box Plot of values by category')
plt.xlabel('category')
plt.ylabel('value')
plt.grid(True)
plt.show()
```

```python
#violin Plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='category', y='values', data=data, palette='Set3')
plt.title('violin Plot of values by category')
plt.xlabel('category')
plt.ylabel('value')
plt.grid(True)
plt.show()
```

```python
#count Plot
plt.figure(figsize=(10, 6))
sns.countplot(x='category', data=data, palette='Set1')
plt.title('count Plot of values by category')
plt.xlabel('category')
plt.ylabel('count')
plt.grid(axis='y')
plt.show()
```

```python
#data set
data=pd.DataFrame(
    {
        'A' : np.random.rand(50),
        'B' : np.random.rand(50),
        'C' : np.random.rand(50),
        'D' : np.random.rand(50)
    }
)
```

```python
#Correlation HeatMap
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation HeatMap')
plt.show()
```

```python
#sample data
np.random.seed(42)
x= np.random.rand(100)*100
y=2*x + np.random.normal(0, 25, 100)
```

```python
#Regression Plot
plt.figure(figsize=(10, 6))
sns.regplot(x = x, y = y, color='b')
plt.title('Regression Plot of x vs y')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
#sample data
data=sns.load_dataset('iris')

#pair plot
sns.pairplot(data, hue='species', palette='Set1')
plt.show()
```

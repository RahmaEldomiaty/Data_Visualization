```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sample data
dates= pd.date_range(start='2023-01-01', periods=12, freq='M')
sales=[15, 25, 32, 40, 22, 17, 30, 18, 36, 44, 38, 50]
Budget=[12, 18, 25, 28, 30, 40, 33, 38, 44, 50, 42, 55]
dates, sales, Budget
```

```python
#create subplot with different sizes
fig= plt.figure(constrained_layout=True, figsize=(14,10))
gs = fig.add_gridspec(3, 3)

ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(dates, sales, marker='o', color= 'b')
ax1.set_title('Sales over Time')

ax2 = fig.add_subplot(gs[0, 2])
ax2.bar(['A', 'B','C','D'], [20, 35, 30, 35])
ax2.set_title('Category Comparison')

ax3 = fig.add_subplot(gs[1:, 1:])
ax3.scatter(Budget, sales, color='r')
ax3.set_title('Advertising Budget vs Sales')

ax4 = fig.add_subplot(gs[1:, 0])
ax4.plot(dates, Budget, marker='s', color= 'g', linestyle='--')
ax4.set_title('Advertising Budget over Time')

plt.show()
```
